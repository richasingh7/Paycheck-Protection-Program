```python
# Importing neccessary libraries
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns

# Ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading the data and saving into a dataframe
df = pd.read_csv('public_150k_plus_230101.csv')   # Loans above 150k
df2 = pd.read_csv('public_up_to_150k_1_230101.csv') # Loans below 150k
```

```python
# Joining that means concatinating the data
result = pd.concat([df,df2])
```

```python
n_df['loanrange']=n_df['loanrange'].astype(int)
```

```python
dff = n_df.groupby(["Race"]).loanrange.sum().reset_index()
```

```python
#Getting info
dff
```

```python
#defining function
def small_chart(plot):
    plot.rcParams.update(plt.rcParamsDefault)
```

# Visualisation

```python
#Importing library
import seaborn as sns
small_chart(plt)
sns.barplot('loanrange', 'Race', data=dff)
plt.xticks(rotation=90)
plt.title('loan range vs race of borrowers')
plt.xlabel('Loan Range')
plt.ylabel('race')
plt.show()
```
